```python
from Crypto.Cipher import AES
from utils import read_file, write_to_file
from Crypto.Random import get_random_bytes

# Generate a random 128-bit AES key
def generate_key(key_file):
    key = get_random_bytes(16)
    write_to_file(key_file, key)
    return key


# Load AES key from a file
def load_key(key_file):
    key = read_file(key_file)
    return key


# Encrypt a file using AES encryption
def encrypt_file(input_file, encrypted_file, key):

    plaintext = read_file(input_file)
    
    cipher = AES.new(key, AES.MODE_EAX)
    
    ciphertext, tag = cipher.encrypt_and_digest(plaintext)
    
    write_to_file(encrypted_file, (cipher.nonce + tag + ciphertext))


# Decrypt a file using AES decryption.
def decrypt_file(encrypted_file, decrypted_file, key):

    data = read_file(encrypted_file)
    nonce, tag, ciphertext = [ data[0:16], data[16:32], data[32:] ] 

    cipher = AES.new(key, AES.MODE_EAX, nonce)

    plaintext = cipher.decrypt_and_verify(ciphertext, tag)

    write_to_file(decrypted_file, plaintext)



# Demo Usage
if __name__ == "__main__":

    # Files paths
    key_file = "AES Demo Files/key.pem"
    input_file = "AES Demo Files/email.txt"
    encrypted_file = "AES Demo Files/email.enc"
    decrypted_file = "AES Demo Files/email.dec"

    # Get AES key from file
    aes_key = load_key(key_file)

    # Encrypt the file
    encrypt_file(input_file, encrypted_file, aes_key)

    # Decrypt the file
    decrypt_file(encrypted_file, decrypted_file, aes_key)

    print(f"File encryption and decryption completed with AES key: 0x{aes_key.hex()}")
```
